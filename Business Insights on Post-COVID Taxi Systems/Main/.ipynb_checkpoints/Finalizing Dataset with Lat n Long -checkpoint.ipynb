```python
import pandas as pd
```

```python
df1 = pd.read_csv('./Final_dataset.csv')
df1.head()
```

```python
df2 = pd.read_csv('./Taxi_Zone_final.csv')
df2.head()
```

```python
df3 = df3.drop(['Unnamed: 0', 'X', 'Y', 'OBJECTID', 'Shape_Leng', 'Shape_Area', 'zone','borough', 'Within NYC or Not'], axis=1)
df3 = df3.rename(columns={"LocationID":"PULocationID", "latitude":"PUlatitude", "longitude":"PUlongitude"})
df3.head()
```

```python
df4 = df2.drop(['Unnamed: 0', 'X', 'Y', 'OBJECTID', 'Shape_Leng', 'Shape_Area', 'zone','borough', 'Within NYC or Not'], axis=1)
df4 = df4.rename(columns={"LocationID":"DOLocationID", "latitude":"DOlatitude", "longitude":"DOlongitude"})
df4.head()
```

```python
df5 = pd.merge(df1, df3, how="left", on=["PULocationID", "PULocationID"])
df5.head(20)
```

```python
df6 = pd.merge(df5, df4, how="left", on=["DOLocationID", "DOLocationID"])
df6.head(20)
```
