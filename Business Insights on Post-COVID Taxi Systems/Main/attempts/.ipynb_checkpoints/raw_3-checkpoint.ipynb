#### Demographic information of Customer & Vendor

    1. 'VendorID' – code indicating the TPER ( Taxicab and Livery Passeng  er ENchancement Programs (TPEP ) provider that provided the record	
	2. 'tpep_pickup_datetime' – pick up date and time 
	3. 'tpep_dropoff_datetime' – drop off date and time
    4. 'passenger_count' – passenger count 
	5. 'trip_distance' – elapsed trip distance in miles reported by the taximeter 
	6. 'RatecodeID' – final rate code effect at the end of the trip
		1 – Standard Rate
		2 – JFK
		3 – Newark
		4 – Nassau
		5 – Negotiated fare
		6 – group ride
	7. 'store_and_fwd_flag' – flag indicated whether the trip record was held in vehicle
		Y – store and forward trip
		N – not a store and forward trip
    8. 'PULocationID' – TLC Taxi zone in which taximeter was engaged 
	9. 'DOLocationID' – TLC Taxi zone in whoch taximeter was disengaged 
	10. 'payment_type' – numeric code signifying the payment mode
		1 – credit card
		2 – cash
		3 – no charge
		4 – dispute
		5 – unknown
		6 – voided trip
 	11. 'fare_amount' – time and distance fare calculated by the meter 
	12. 'extra' – miscellaneous extras and surcharges.
		1. Rush charge – 0.5 USD
		2. Overnight charge – 1 USD
    13. 'mta_tax' – 0.50 MTA tax, automatically added 
	14. 'tip_amount' – tip amount. field is automatically populated for credit card tips., cash tips are not included. 
	15. 'tolls_amount' – total amount of all tolls paid in trip 
	16. 'improvement_surcharge' – 0.30 imporvement surcharge assessed trips at the flag drop.
        17. 'total_amount' – total amount charged to passengers. Does not include cash tips. 
	18. 'congestion_surcharge' – additional charge added to base charge during unusal events like strikes, bad winter, major port fires.
    19. zone_x - pick up zone
    20. borough_x - pick up borough
    21. PUlatitude - pick up latitude
    22. PUlongitude - pick up longitude
    23. zone_y - drop off zone
    24. borough_y - drop off borough
    25. DOlatitude - drop off latitude
    26. DOlongitude - drop off longitude

#### Separate the independent and dependent relationship

```python
x = df.drop(['VendorID', 'tpep_pickup_datetime', 'tpep_dropoff_datetime',
       'passenger_count', 'RatecodeID', 'store_and_fwd_flag',
       'PULocationID', 'DOLocationID', 'payment_type', 'fare_amount', 'extra',
       'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge',
       'total_amount', 'congestion_surcharge', 'zone_x', 'borough_x',
       'PUlatitude', 'PUlongitude', 'zone_y', 'borough_y', 'DOlatitude',
       'DOlongitude', 'day_of_week', 'pickup_hour', 'dropoff_hour',
       'day_of_month', 'ride_duration_min'], axis=1).values
```

```python
y = df['fare_amount']
```

#### Split the data into train and tests sets

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, random_state=1234567)
```

#### Create and train the model

```python
def lg():
    from sklearn.linear_model import LogisticRegressionCV
    
    model = LogisticRegressionCV()
    
    params = model.get_params()
    
    model.fit(x_train, y_train)
    
    return model
```

```python
def knn():
    from sklearn.neighbors import KNeighborsRegressor
    
    model = KNeighborsRegressor()
    
    params = model.get_params()
    
    model.fit(x_train, y_train)
    
    return model
```

```python
def svm():
    from sklearn.svm import SVR
    
    model = SVR(C=2)
    
    params = model.get_params()
    
    model.fit(x_train, y_train)
    
    return model
```

```python
def decision_tree():
    from sklearn.tree import DecisionTreeRegressor
    
    model = DecisionTreeRegressor()
    
    params = model.get_params()
    
    model.fit(x_train, y_train)
    
    return model
```

```python
def random_forest():
    from sklearn.ensemble import RandomForestRegressor
    
    model = RandomForestRegressor(n_estimators=100)
    
    params = model.get_params()
    
    model.fit(x_train, y_train)
    
    return model
```

```python
def ada_boost():
    from sklearn.ensemble import AdaBoostRegressor
    
    model = AdaBoostRegressor()
    
    params = model.get_params()
    
    model.fit(x_train, y_train)
    
    return model
```

```python
def gradient_boost():
    from sklearn.ensemble import GradientBoostingRegressor
    
    model = GradientBoostingRegressor()
    
    params = model.get_params()
    
    model.fit(x_train, y_train)
    
    return model
```

```python
def xg_boost():
    from xgboost import XGBRegressor
    
    model = XGBRegressor()
    
    params = model.get_params()

    model.fit(x_train, y_train)
    
    return model
```

```python
def linear_regression():
    from sklearn.linear_model import LinearRegression

    # create the model
    model = LinearRegression()
    
    # train the model
    model.fit(x_train, y_train)
    
    return model
```

```python
def elastic_net():
    from sklearn.linear_model import ElasticNet

    # create the model
    model = ElasticNet()
    
    # train the model
    model.fit(x_train, y_train)
    
    return model
```

```python
def lasso():
    from sklearn.linear_model import Lasso

    # create the model
    model = Lasso()
    
    # train the model
    model.fit(x_train, y_train)
    
    return model
```

```python
def ridge():
    from sklearn.linear_model import Ridge

    # create the model
    model = Ridge()
    
    # train the model
    model.fit(x_train, y_train)
    
    return model
```

#### model evaluation

```python
def evaluate_model(model, name):
    
    from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

    predictions = model.predict(x_test)
    predictions

    mae = mean_absolute_error(y_test, predictions)

    mse = mean_squared_error(y_test, predictions)

    rmse = np.sqrt(mse)

    r2 = r2_score(y_test, predictions)

    result = {
        "mae" : mae,
        "mse" : mse,
        "rmse" : rmse,        
        "r2" : r2
    }
    
    return result
```

```python
results = []

results.append(evaluate_model(linear_regression(), 'Linear Regression'))
results.append(evaluate_model(elastic_net(), 'Elastic Net Regression'))
results.append(evaluate_model(lasso(), 'Lasso Regression'))
results.append(evaluate_model(ridge(), 'Ridge Regression'))
# results.append(evaluate_model(lg(), 'Logistic Regression'))
results.append(evaluate_model(knn(), 'KNN'))
results.append(evaluate_model(svm(), 'SVM'))
results.append(evaluate_model(decision_tree(), 'Decision Tree'))
results.append(evaluate_model(random_forest(), 'Random Forest'))
results.append(evaluate_model(ada_boost(), 'Ada Boost'))
results.append(evaluate_model(gradient_boost(), 'Gradient Boost'))
results.append(evaluate_model(xg_boost(), 'Extreme Gradient Boost'))
```

```python
algorithms = ['Linear Regression', 'Elastic Net Regression', 'Ridge Regression', 'Lasso Regression', 'KNN', 'SVM', 'Decision Tree', 'Random Forest', 'Ada Boost', 'Gradient Boost', 'XG Boost']
# algorithms = ['Elastic Net Regression', 'Ridge Regression', 'Lasso Regression']
algo_insights = pd.DataFrame(results, index= algorithms)
algo_insights.columns = ["mae", "mse", "rmse", "r2"]
algo_insights.sort_values(by="r2", ascending=False)
# print(algo_insights)
```
